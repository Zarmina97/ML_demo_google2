## Summary

Let's analyse just the df_train dataset for now.
Some notes:
- User ID 1001680 has the highest purchase count in the dataset, whereas User ID 1000708 has the lowest.
- Male customers have a higher purchase volume and also contribute to a greater financial expenditure overall.
- The 26-35 age bracket shows the highest quantity of items purchased, yet the expenditure is relatively uniform across all age groups.
- Individuals with occupation code 4 stand out for purchasing significantly more items, though the spending is fairly consistent across different occupations.
- City category B leads in the number of items purchased, while category C exhibits higher spending.
- Residents with a tenure of 1 year show the highest number of items purchased, but the spending is comparable up to a tenure of 5+ years.
- Single individuals make more purchases, but the expenditure is comparable to that of individuals who are married.

## Gender

```python
%matplotlib inline
percentages = (df_train['Gender'].value_counts() / df_train['Gender'].value_counts().sum()) * 100
counts = df_train['Gender'].value_counts()
plt.figure(figsize=(10, 6))
bars=sns.barplot(x=counts.index, y=counts.values, order=list(counts.index))
plt.title("Percentage of Men & Women who Purchased more items", fontsize="15")
# Aggiungi le percentuali su ogni barra
for bar, percentage in zip(bars.patches, percentages):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2., height,
                 f'{percentage:.1f}%',  # Formattazione con una cifra decimale
                 ha='center', va='bottom', fontsize=10)
plt.xlabel("Gender", fontsize="12")
plt.ylabel("Counts", fontsize="12")
plt.xticks(rotation=0)
# Rimuovi gli spigoli del grafico
sns.despine()
plt.savefig('res/data exploration/gender_n_purchases.png')
plt.show()
```

```python
%matplotlib inline
avg_purchase_ = df_train.groupby('Gender').agg(avg_purchase=('Purchase', np.mean))
total_avg_purchase = avg_purchase_['avg_purchase'].sum()

percentages = (avg_purchase_['avg_purchase'] / total_avg_purchase) * 100
plt.figure(figsize=(10, 6))
bars=sns.barplot(x=avg_purchase_.index, y=avg_purchase_['avg_purchase'])
plt.title("Percentage of Men & Women who spend more", fontsize="15")

for bar, percentage in zip(bars.patches, percentages):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2., height,
                 f'{percentage:.3f}%',  # Formattazione con una cifra decimale
                 ha='center', va='bottom', fontsize=10)
        
plt.xlabel("Gender", fontsize="12")
plt.ylabel("Counts", fontsize="12")
plt.xticks(rotation=0)
sns.despine()
plt.savefig('res/data exploration/gender_avg_purchases.png')
plt.show()
```

**<code>Product_Category_1</code> and <code>Purchase</code>**: The moderate negative correlation of -0.34 between <code>Product_Category_1</code> and <code>Purchase</code> suggests that as the product category 1 increases, the purchase amount tends to decrease. This could imply that higher category numbers are associated with less expensive products, or perhaps these products are less popular or purchased less frequently. This information could be used to analyze customer purchasing patterns or to adjust inventory and marketing strategies.

**<code>Product_Category_3</code> and <code>Purchase</code>**: The positive correlation of 0.29 between <code>Product_Category_3</code> and <code>Purchase</code> might indicate that products in this category are more expensive or that when they are bought, they are part of larger transactions. It suggests that products in category 3 could be driving higher sales values. This might be an area for businesses to focus on, such as promoting these products or bundling them with others.

**Correlations Close to Zero**: Several variables such as <code>Gender</code>, <code>Age</code>, <code>Occupation</code>, <code>City_Category</code>, <code>Stay_In_Current_City_Years</code>, and <code>Marital_Status</code> have correlations close to zero with Purchase, suggesting that these factors do not have a strong linear relationship with the amount spent. However, this doesn't mean they are not important; they could still influence purchasing behavior in non-linear ways or through interactions with other variables.

**Potential for Segmentation**: Although some demographic variables like Gender and Age show a low correlation with Purchase, they can still be valuable for market segmentation and targeted marketing campaigns. For instance, if gender has a correlation, even if small, businesses might find that certain products are slightly more popular with one gender over the other.

**Data-Driven Decisions**: The correlation information can help businesses make data-driven decisions about which product categories to promote or discount, which demographic segments to target, and how to forecast sales.

**Consideration for Marketing and Stocking**: If certain product categories are shown to have a stronger relationship with purchases, marketing efforts could be tailored to highlight these categories, and stock levels could be adjusted accordingly to ensure supply meets potential demand.

**Pricing Strategies**: The negative correlation with <code>Product_Category_1</code> might suggest a need to review pricing strategies for these products or to consider cross-selling opportunities with other product categories that show a positive correlation with Purchase.

Correlation does not imply causation. Therefore, while the correlation matrix can provide insights, it's important to conduct further analysis, such as regression modeling or hypothesis testing, to understand the causal relationships and how different variables might affect purchasing behavior.
